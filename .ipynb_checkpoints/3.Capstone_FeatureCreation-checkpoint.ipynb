### 2.3 Feature Engineering

Already in the initial data exploration, the correlation between features was observed. Let's compute it again for our cleansed dataset.

```python
import pandas as pd

data = pd.read_csv('cleansed_data.csv')
data = data.drop(['Unnamed: 0'], axis=1)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = data.drop(['diagnosis'], axis=1).corr()
plt.subplots(figsize=(15,12))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

From here we learn that there are groups of correlated features:
- *compactness mean, smoothness mean concavity mean* and *concave points mean*. I took *concave points mean*
- *radius se, perimeter se* and *area se*. I took *area_se*
- *radius worst, perimeter worst* and *area worst*. I took *area_worst*
- *compactness worst, concavity worst* and *concave points worst*. I took *concave points worst*
- *compactness se, concavity se* and *concave points se*. I took *concave points se*
- *texture mean* and *texture worst*. I took *texture worst*
- *area worst*, *radius mean*, *perimeter mean* and *area mean*. I took *area mean*

```python
data.columns
```

```python
droplist = ['compactness_mean', 'concavity_mean', 'radius_se', 'perimeter_se',
           'radius_worst', 'perimeter_worst', 'area_worst', 'compactness_worst', 
           'concavity_worst', 'compactness_se', 'concavity_se',
           'texture_mean','area_worst', 'radius_mean', 'perimeter_mean', 'concave points_worst']
data_filt = data.drop(droplist, axis=1)
data_filt.head()
```

```python
corr_matrix = data_filt.drop(['diagnosis'], axis=1).corr()
plt.subplots(figsize=(15,12))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
