### 2.2 Data Cleansing

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualization library  
import matplotlib.pyplot as plt
#https://jupyterlab-9.labs.cognitiveclass.ai/hub/user-redirect/lab/tree/labs/User/Advanced%20IBM%20Data%20Science%20/Capstone/data.csv
```

```python
data_raw = pd.read_csv('data.csv')
data = data_raw.drop(['id', 'Unnamed: 32'], axis=1)
data.describe()
```

I converted the target column to categorical binary data.

```python
data['diagnosis'] = data['diagnosis'].astype('category')
data.dtypes
```

```python
dataviz = pd.melt(data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=dataviz,split=True, inner="quart")
plt.xticks(rotation=90)
```

In this case, features contain values in a wide variety of ranges. For this reason, normalization was performed.

```python
from sklearn import preprocessing
label = data.diagnosis
unlabeled_data = data.drop(['diagnosis'],axis = 1 )

x = unlabeled_data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
unlabeled_data_scaled = pd.DataFrame(x_scaled)
unlabeled_data_scaled.columns = unlabeled_data.columns
data_scaled = pd.concat([label, unlabeled_data_scaled], axis=1)
data_scaled.head()
```

```python
dataviz = pd.melt(data_scaled,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=dataviz,split=True, inner="quart")
plt.xticks(rotation=90)
```

#### Outliers detection
For the identification of outliers, Z-score was used. The term Z-score, standardized unit, reduced center variable, standardized or normalized variable is used in statistics to compare data from different samples or populations and is defined as the number of standard deviations that a given value takes from the mean of its sample or population.

For each feature, Z-score is computed, relative to that column mean and sd. Then, the absolute of Z-score is taken because the sign does not matter, only if it's below the threshold. Then it is applied to every element in a row.

```python
data_scaled.shape
```

```python
from scipy import stats
data_scaled_o = data_scaled[(np.abs(stats.zscore(data_scaled.iloc[:,1:31])) < 3).all(axis=1)]
data_scaled_o.shape
```

We have herby removed 74 samples that could be considered outliers for some of the studied features.

```python
data_scaled_o.to_csv("cleansed_data.csv")
```
